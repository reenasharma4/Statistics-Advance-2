#1. What is hypothesis testing in statistics?
  - Hypothesis testing is just a structured way of making decisions based on data. It involves testing an assumption about the population parameter and determine on the basis of that whether there is enough evidence to reject or fail to reject that assumption.

#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - Null Hypothesis (H0): This is a "nothing special" hypothesis or it represents the default. It says that there is no difference, no change, no effect or relationship between the variables.

  Example: A new weight loss pill does not help people to lose their weight.

  - Alternative Hypothesis (H1 or H1): It is the opposite of the null hypothesis. It suggest that something is happening and says that there is a difference, change, effect or relationship.

  Example: The new weight loss pill does help people to lose their weight.

#3. What is the significance level in hypothesis testing, and why is it important?
  - It is a threshold for determining whether to reject the null hypothesis or not. It represents the possibility of making the wrong decision or type I error.
  - Common significance levels are 0.05 (5%) or 0.01 (1%).
  - A lower Œ± reduces the risk of a type error I but increases the risk of a type error II.

  The significance level helps define the cutoff point for decision-making in hypothesis testing.

#4. What does a P-value represent in hypothesis testing?
  - In hypothesis testing, the P-value is a measure that helps to determine the strength of evidence against the null hypothesis (H0). It represents the probability of obtaining the observed results (or more extreme ones) if H0 were true.
  - Interpretation of P-Value:
    * Small P-value (P<0.05): Strong evidence against H0, suggesting that we reject the null hypothesis and accept the alternative hypothesis (H1).
    * Large P-value (P>0.05): Weak evidence against H0, meaning we do not have enough proof to reject the null hypothesis.

#5. How do you interpret the P-value in hypothesis testing?
  - If P‚â§Œ± (e.g., 0.05 or 0.01): Reject H0. The result is statistically significant, meaning there is enough evidence to support H1.
  - If P>Œ±: Fail to reject H0. There is not enough evidence to support H1, meaning we do not have strong proof that the null hypothesis is false.

  Example Interpretation:
    * A P-value of 0.03 (with Œ±=0.05) ‚Üí Reject H0, significant result.
    * A P-value of 0.08 (with Œ±=0.05) ‚Üí Fail to reject H0, not significant.

#6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Imagine you are testing whether a medicine works.
    1. Type 1 Error (False Positive): Saying the medicine works when it actually does not.
    2. Type 2 Error (False Negative): Saying the medicine does not work when it actually does.

In simpler terms, Type 1 error is like a false alarm, and Type 2 error is like missing a real issue.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   - One-tailed test: Checks if something is either greater or lesser than a certain value (not both). For Example: Checking if a new diet helps people lose more weight than before.
   - Two-tailed test: Checks if something is different (either greater or lesser). For Example: Checking if a new drug affects blood pressure either positively or negatively.

#8. What is the Z-test, and when is it used in hypothesis testing?
  - A Z-test is used when we compare a sample to a known population and we know the standard deviation. For Example: If a company claims the average salary of employees is $50,000, we use a Z-test to check if our sample data supports this claim.

#13. What is the relationship between Z-test and T-test in hypothesis testing?

#14. What is a confidence interval, and how is it used to interpret statistical results?
   - A confidence interval gives us a range where the true value is likely to fall. For Example: If we say the average height is 170 cm ¬± 5 cm (95% confidence), we are 95% sure the true average is between 165 cm and 175 cm.

#15. What is the margin of error, and how does it affect the confidence interval?

       * Example: Margin of error = 2 cm ‚Üí Average height = 170 ¬± 2 cm
       * Margin of error = 5 cm ‚Üí Average height = 170 ¬± 5 cm

#16. How is Bayes' Theorem used in statistics, and what is its significance?

#17. What is the Chi-square distribution, and when is it used?

#18. What is the Chi-square goodness of fit test, and how is it applied?

#19. What is the F-distribution, and when is it used in hypothesis testing?
   - The F-distribution is used to compare the variances of two datasets. For Example: Checking if the performance of the students in two different schools has the same variation in marks.

#20. What is an ANOVA test, and what are its assumptions?
   - ANOVA (Analysis of Variance) is used to compare the three or more groups. For Example: Comparing the average test scores of students in three different schools.
   - Assumptions of ANOVA:
     1. Data follows a normal distribution.
     2. The samples are independent.
     3. The variances of the groups should be equal.

#21. What are the different types of ANOVA tests?

   - Two-way ANOVA: Compares two factors at once. For Example: Checking how gender and study habits affect test scores.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    if p_value < alpha:
        result = "Reject the null hypothesis: The sample mean significantly differs from the population mean."
    else:
        result = "Fail to reject the null hypothesis: No significant difference between sample and population means."

    return z_score, p_value, result

sample_data = [100, 102, 98, 97, 101, 99, 103, 105, 96, 100]
population_mean = 98
population_std = 3

z_score, p_value, interpretation = z_test(sample_data, population_mean, population_std)
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
from statsmodels.stats.weightstats import ztest

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Sample drawn from a normal distribution

pop_mean = 50  # Hypothesized population mean

z_stat, p_value = ztest(sample, value=pop_mean)

print(f"Z-statistic: {z_stat}, P-value: {p_value}")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

alpha = 0.05

z_critical = stats.norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 5))
sns.lineplot(x, y, label="Standard Normal Distribution")

plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Rejection Region")

z_stat, _ = ztest(sample, value=pop_mean)

plt.axvline(z_stat, color='blue', linestyle='dashed', label=f"Sample Z = {z_stat:.2f}")

plt.axvline(-z_critical, color='red', linestyle='dotted', label=f"Critical Z = {-z_critical:.2f}")
plt.axvline(z_critical, color='red', linestyle='dotted', label=f"Critical Z = {z_critical:.2f}")

plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(8, 5))
    sns.lineplot(x, y, label="Standard Normal Distribution")

    alpha = 0.05
    z_critical = stats.norm.ppf(1 - alpha/2)

    # Shade Type 1 error region
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Type 1 Error (Œ±)")

    # Shade Type 2 error region
    beta_region = (-1.5, 1.5)
    plt.fill_between(x, y, where=(x > beta_region[0]) & (x < beta_region[1]), color='blue', alpha=0.3, label="Type 2 Error (Œ≤)")

    plt.axvline(-z_critical, color='red', linestyle='dotted', label=f"Critical Z = {-z_critical:.2f}")
    plt.axvline(z_critical, color='red', linestyle='dotted', label=f"Critical Z = {z_critical:.2f}")

    plt.legend()
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.show()

visualize_type_errors()
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
import scipy.stats as stats

def independent_t_test():
    np.random.seed(42)
    sample1 = np.random.normal(50, 10, 30)  # Random sample
    sample2 = np.random.normal(55, 10, 30)  # Another random sample

    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    print(f"Independent T-test: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

independent_t_test()
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.
import seaborn as sns
import matplotlib.pyplot as plt

def paired_t_test():
    np.random.seed(42)
    sample1 = np.random.normal(50, 10, 30)
    sample2 = sample1 + np.random.normal(2, 5, 30)  # Paired sample with slight variation

    t_stat, p_value = stats.ttest_rel(sample1, sample2)
    print(f"Paired T-test: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

    sns.histplot([sample1, sample2], kde=True)
    plt.title("Paired Sample Comparison")
    plt.show()

paired_t_test()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval():
    np.random.seed(42)
    sample = np.random.normal(50, 10, 100)

    mean = np.mean(sample)
    sem = stats.sem(sample)
    interval = stats.t.interval(0.95, len(sample)-1, loc=mean, scale=sem)

    print(f"95% Confidence Interval: {interval}")

confidence_interval()
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error():
    np.random.seed(42)
    sample = np.random.normal(50, 10, 100)

    sem = stats.sem(sample)
    margin = sem * stats.t.ppf((1 + 0.95) / 2., len(sample)-1)

    print(f"Margin of Error: {margin:.4f}")

margin_of_error()
```

```python
#16. D Implement an F-test using Python to compare the variances of two random samples.
def f_test():
    np.random.seed(42)
    sample1 = np.random.normal(50, 10, 30)
    sample2 = np.random.normal(50, 15, 30)

    f_stat = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)
    p_value = stats.f.cdf(f_stat, len(sample1)-1, len(sample2)-1)

    print(f"F-test: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

f_test()
```

```python
#18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
def two_way_anova():
    data = pd.DataFrame({
        "Factor1": np.tile(["X", "Y"], 15),
        "Factor2": np.repeat(["A", "B", "C"], 10),
        "Score": np.random.normal(50, 10, 30)
    })

    model = ols('Score ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(anova_table)

    sns.boxplot(x="Factor1", y="Score", hue="Factor2", data=data)
    plt.title("Two-Way ANOVA")
    plt.show()

two_way_anova()
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_column, value_column):
    groups = [group[value_column].values for _, group in data.groupby(group_column)]

    # 1. Normality check (Shapiro-Wilk test)
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test: p-value = {p:.4f}")

    # Q-Q Plot
    plt.figure(figsize=(6,4))
    for i, group in enumerate(groups):
        stats.probplot(group, dist="norm", plot=plt)
    plt.title("Q-Q Plot for Normality Check")
    plt.show()

    # 2. Homogeneity of variance (Levene‚Äôs Test)
    stat, p = stats.levene(*groups)
    print(f"Levene‚Äôs Test for Equal Variance: p-value = {p:.4f}")

    # 3. Boxplot for variance comparison
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[group_column], y=data[value_column])
    plt.title("Boxplot for Variance Comparison")
    plt.show()
```
